```python
def count_employees_test():
    # Locate the <h2> element by its class name
    element = driver.find_element(By.CLASS_NAME, "text-heading-xlarge")
    
    # Extract the text from the element
    text = element.text
    
    # Use regex to extract the number as an integer
    number = int(re.search(r'\d+', text).group())
    return number
```

```python
from selenium.webdriver.common.by import By
import re

def count_associated_members():
    # Locate the <h2> element by its CSS selector
    element = driver.find_element(By.CSS_SELECTOR, ".org-people__header-spacing-carousel h2")
    
    # Extract the text from the element
    text = element.text
    
    # Use regular expressions to find the integer in the string
    number = int(re.search(r'\d+', text).group())
    
    #print(number)  # This will print the number 5 as an integer
    return number
```

```python
# Loop through each row in the DataFrame-> all functions out to the connected notebooks will be called in this loop
i=0
for index, row in input.iterrows():
    i=i+1
    print(i)
    try:
        # Extract values from the current row
        internal_id = row['internal_id_x']
        company_name = row['Company Name']
        linkedin_url_x = row['linkedin_url_x']
        employee_range = row['Employee Range']
        print(employee_range)
        
        # Extract Employee Data
        dataframes_Persons_Statistics = Masterthesis_Persons_Statistics.persons_statistics(internal_id, linkedin_url_x, dataframes_Persons_Statistics, driver)
        #dataframes_Persons_Statistics[f"df_1_internal_id"]

        List = ['11-50 employees', '101-250 employees', '251-500 employees', '51-100 employees']

        try:
            employee_number = count_employees_test()
            if employee_number > 10:
                Connector.Founder_Filter(driver)
        except:
            try:
                employee_number = count_associated_members()
                if employee_number > 10:
                    Connector.Founder_Filter(driver)
                    print("bnane")
            except:
                if employee_range in List:
                    Connector.Founder_Filter(driver)
    
        # Extract Connector DataFrame
        counts_df, linkedin_links_df = Connector.Connect(internal_id, company_name, counts_df, linkedin_links_df, driver)
        
        # Navigate to Insights
        Insights.To_Insights(driver)
        
        #Scroll Load
        Connector.ScrollAndLoad(driver)
        # Extract insights Statistics
        employee_number_df = Insights.employee_counter(internal_id, company_name, employee_number_df, driver)
        employee_diversity_df = Insights.diversity(internal_id, company_name, employee_diversity_df, driver)
    except:
        pass
```
