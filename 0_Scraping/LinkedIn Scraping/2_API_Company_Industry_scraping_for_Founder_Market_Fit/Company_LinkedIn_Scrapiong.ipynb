```python
import pandas as pd
from linkedin_api import Linkedin
import json
import time
import random

# Load the CSV file
df = pd.read_csv(r"unique_companies.csv")

# Extract the LinkedIn usernames from the URLs
df['company_id'] = df['company_link'].str.extract(r'company/([^/]+)')

# Authenticate with LinkedIn
api = Linkedin("email", "password")#,refresh_cookies=True)


def fetch_company_data(api, company_id):
    try:
        company_data = api.get_company(company_id)
        #print(company_data)
        return company_data
    except Exception as e:
        print(f"Error fetching company for {company_id}: {e}")
        return None

company_data_list = []

error_companies = []
interaction_count = 3639

# JSON file path
json_file_path = r'C:..\LinkedIn_API_Company_data\unique_companies_3640_end.json'

# Create a new JSON file or overwrite if it exists
with open(json_file_path, 'w', encoding='utf-8') as json_file:
    json.dump([], json_file)  # Initialize with an empty list

# Process each company
for company_name in df['company_id'].iloc[3640:]:
    company_data = fetch_company_data(api, company_name)  # Fetch data using the provided API
    interaction_count += 1
    print(f'Data Fetched: {interaction_count}')
    
    if company_data:
        # Append new data to the JSON file incrementally
        with open(json_file_path, 'r+', encoding='utf-8') as json_file:
            existing_data = json.load(json_file)  # Load existing data
            existing_data.append(company_data)  # Append new company data
            json_file.seek(0)  # Move to the start of the file
            json.dump(existing_data, json_file, indent=4, ensure_ascii=False)  # Write updated data
    else:
        company_link = f"https://www.linkedin.com/in/{company_name}"
        error_companies.append(company_link)

    # Introduce a random delay between 3 and 10 seconds
    time.sleep(random.uniform(3, 10))

# Print the profiles that encountered errors
print("Companies that encountered errors:")
for company_link in error_companies:
    print(company_link)
```
