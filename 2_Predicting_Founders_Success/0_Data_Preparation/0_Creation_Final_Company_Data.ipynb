### Extract start and end date of the company from the founder experience and drop the companies who's founders don't have the company in the experience
 -  They don't take it seriously enough and I could not extract the start and end date for these companies out of the experience section

```python
import pandas as pd
import ast
from difflib import SequenceMatcher

# Function to match company names
def is_match(org_name, exp_company_name, threshold=0.85):
    # Check if either name is in the other, ignoring case
    if org_name.lower() in exp_company_name.lower() or exp_company_name.lower() in org_name.lower():
        return True
    # Use SequenceMatcher for partial matching
    return SequenceMatcher(None, org_name.lower(), exp_company_name.lower()).ratio() >= threshold

# Initialize matched and unmatched DataFrames
matched_rows = []
trash_rows = []

# Iterate through each row of merged_df
for idx, row in merged_df.iterrows():
    org_name = row['Organization Name'].split(';')[0]  # Extract the company name
    try:
        # Parse the experience column string into a list of dictionaries
        experiences = ast.literal_eval(row['experience'])
    except (ValueError, SyntaxError):
        # If parsing fails, move the row to trash_df
        trash_rows.append(row)
        continue

    # Flag to indicate if a match was found
    found_match = False

    for exp in experiences:
        exp_company_name = exp['company_name']
        # Check if the organization name matches the experience company name
        if is_match(org_name, exp_company_name):
            # Add start_date and end_date columns for the matched company
            row['start_date'] = exp.get('start_date')
            row['end_date'] = exp.get('end_date')
            matched_rows.append(row)
            found_match = True
            break

    if not found_match:
        trash_rows.append(row)

# Create DataFrames for matched and unmatched rows
matched_df = pd.DataFrame(matched_rows)
trash_df = pd.DataFrame(trash_rows)

# Display results
#print("Matched DataFrame:")
#print(matched_df)
#print("\nTrash DataFrame:")
#print(trash_df)

matched_df
```

Ensure consistency of start end end date. If there are several founders some might have joined in earlyer or didn't have stayed until the end. But the earliest start and the latest end value should show the beginnign and end of the company.

```python
import pandas as pd
from datetime import datetime
import numpy as np

# Helper function to parse dates consistently
def parse_date(date_str):
    try:
        if len(date_str.strip()) == 4 and date_str.isdigit():  # Handles years like "2016"
            return datetime.strptime(date_str, "%Y")
        return datetime.strptime(date_str, "%b %Y")  # Handles formats like "Dec 2014"
    except:
        return np.nan  # If parsing fails, return NaN

# Parse `start_date` and `end_date` into datetime objects
matched_df['start_date'] = matched_df['start_date'].apply(lambda x: parse_date(x) if pd.notna(x) else x)
matched_df['end_date'] = matched_df['end_date'].apply(lambda x: parse_date(x) if pd.notna(x) else x)

# Function to consolidate dates for each company
def consolidate_dates(group):
    group['start_date'] = group['start_date'].min()  # Earliest start_date
    group['end_date'] = group['end_date'].max()  # Latest end_date
    return group

# Group by company name and apply the consolidation function
matched_df = matched_df.groupby('Organization Name', group_keys=False).apply(consolidate_dates)

# Convert dates back to their original format (if needed)
def format_date(date):
    if pd.isna(date):
        return np.nan
    return date.strftime("%b %Y") if isinstance(date, datetime) and date.day == 1 else date.strftime("%Y")

matched_df['start_date'] = matched_df['start_date'].apply(lambda x: format_date(x))
matched_df['end_date'] = matched_df['end_date'].apply(lambda x: format_date(x))

# Display the updated DataFrame
matched_df
```

## Measure how much time passes on average between the last funding rounds before the acquisition and the acquisition to make a valid assumption for the CLosed Dataset, to assume that startups did not raise any money anymore until one year before the acquisition

```python
import pandas as pd

# Convert acquisition_announced_on_date and funding date columns to datetime
date_columns = [
    'date_Pre Seed Round', 'date_Seed Round', 'date_Venture Round',
    'date_Equity Crowdfunding', 'date_Corporate Round', 'date_Convertible Note',
    'date_Private Equity Round', 'date_Series A', 'date_Series B',
    'date_Series C', 'date_Series D', 'date_Series E',
    'date_Series F', 'date_Series G', 'date_Series H', 'date_Post-IPO Equity'
]

# Ensure all date columns and acquisition date are datetime
Open_Startup_Total_df[date_columns + ['acquisition_announced_on_date']] = Open_Startup_Total_df[
    date_columns + ['acquisition_announced_on_date']
].apply(pd.to_datetime, errors='coerce')

# Define a function to calculate the positive time difference
def get_positive_time_difference(row):
    acquisition_date = row['acquisition_announced_on_date']
    # If acquisition date is NaN, return NaN
    if pd.isna(acquisition_date):
        return None

    # Sort funding dates in descending order (most recent first)
    funding_dates = row[date_columns].dropna().sort_values(ascending=False)

    # Iterate through funding dates to find the first positive time difference
    for funding_date in funding_dates:
        time_difference_months = (acquisition_date - funding_date).total_seconds() / (30 * 24 * 60 * 60)  # Convert to months
        if time_difference_months > 0:
            return time_difference_months  # Return positive time difference

    # If no positive time difference is found, return NaN
    return None

# Apply the function to calculate the time differences
Open_Startup_Total_df['time_difference_months'] = Open_Startup_Total_df.apply(get_positive_time_difference, axis=1)

# Copy the column to filtered_df, including NaN values for rows without valid differences
filtered_df = Open_Startup_Total_df[['Company Name', 'time_difference_months']]

# Drop column out of Open_Startup_Total_df
Open_Startup_Total_df = Open_Startup_Total_df.drop(columns=['time_difference_months'])

# Remove rows with NaN values in the time_difference_months column
time_between_funding_and_IPO = filtered_df.dropna(subset=['time_difference_months']).reset_index(drop=True)

# Display the updated DataFrame
from IPython.display import display
display(filtered_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Determine the range of months
min_month = int(time_between_funding_and_IPO['time_difference_months'].min())
max_month = int(time_between_funding_and_IPO['time_difference_months'].max())

# Create bins for each month
bins = np.arange(min_month, max_month + 2)  # +2 to include the last month

# Calculate the average
average = time_between_funding_and_IPO['time_difference_months'].mean()

# Filter for the first 12 months
first_12_months = time_between_funding_and_IPO['time_difference_months'] <= 12
percentage_first_12_months = (first_12_months.sum() / len(time_between_funding_and_IPO)) * 100

# Plot the distribution of time_difference_months
plt.figure(figsize=(10, 6))
plt.hist(
    time_between_funding_and_IPO['time_difference_months'],
    bins=bins,
    edgecolor='black',
    alpha=0.7,
    label='Distribution'
)

# Add a vertical line for the average
plt.axvline(average, color='blue', linestyle='--', linewidth=1.5, label=f'Average: {average:.2f} months')

# Highlight the first 12 months region
plt.fill_betweenx(
    y=[0, plt.gca().get_ylim()[1]],  # Full height of the plot
    x1=min_month,
    x2=12,
    color='green',
    alpha=0.3,
    label=f'First 12 Months: {percentage_first_12_months:.2f}%'
)

# Add labels and title
plt.title('Distribution of Time Between Last Funding Round and Acquisition', fontsize=16)
plt.xlabel('Months', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Add legend
plt.legend(fontsize=12)

# Show grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```python
import pandas as pd

# Function to calculate the new DataFrame
def create_new_dataframe(df):
    # Copy the necessary columns
    new_df = df[['internal_id_x', 'Company Name', 'UUID', 'Founded In']].copy()

    # Convert 'Founded In' to datetime, treating it as a year
    df['Founded In'] = pd.to_datetime(df['Founded In'].astype(str), format='%Y', errors='coerce')
    df['acquisition_announced_on_date'] = pd.to_datetime(df['acquisition_announced_on_date'], errors='coerce')

    # List of columns that represent equity funding
    equity_columns = [
        'US_dollar_Pre Seed Round', 'US_dollar_Seed Round', 'US_dollar_Venture Round',
        'US_dollar_Equity Crowdfunding', 'US_dollar_Corporate Round', 'US_dollar_Convertible Note',
        'US_dollar_Private Equity Round', 'US_dollar_Series A', 'US_dollar_Series B',
        'US_dollar_Series C', 'US_dollar_Series D', 'US_dollar_Series E',
        'US_dollar_Series F', 'US_dollar_Series G', 'US_dollar_Series H', 'US_dollar_Post-IPO Equity'
    ]

    # Corresponding date columns for each equity column
    date_columns = [
        'date_Pre Seed Round', 'date_Seed Round', 'date_Venture Round',
        'date_Equity Crowdfunding', 'date_Corporate Round', 'date_Convertible Note',
        'date_Private Equity Round', 'date_Series A', 'date_Series B',
        'date_Series C', 'date_Series D', 'date_Series E',
        'date_Series F', 'date_Series G', 'date_Series H', 'date_Post-IPO Equity'
    ]

    # Calculate equity_raised_in_5_years
    equity_raised = []
    for index, row in df.iterrows():
        equity_sum = 0
        for equity_col, date_col in zip(equity_columns, date_columns):
            if pd.notna(row[date_col]) and pd.notna(row['Founded In']):
                if (row[date_col] - row['Founded In']).days <= 5 * 365:  # Check if within 5 years
                    equity_sum += row[equity_col]
        equity_raised.append(equity_sum)
    new_df['equity_raised_in_5_years'] = equity_raised

    # Calculate Success_after_5_years
    success = []
    for index, row in df.iterrows():
        if pd.notna(row['acquisition_announced_on_date']) and pd.notna(row['Founded In']):
            if (row['acquisition_announced_on_date'] - row['Founded In']).days <= 5 * 365:  # Check if within 5 years
                success.append(1)
            else:
                success.append(0)
        else:
            success.append(0)
    new_df['Success_after_5_years'] = success

    # Calculate equity_rounds_raised_after_5_years
    equity_rounds_after_5_years = []
    for index, row in df.iterrows():
        count = 0
        for date_col in date_columns:
            if pd.notna(row[date_col]) and pd.notna(row['Founded In']):
                if (row[date_col] - row['Founded In']).days > 5 * 365:  # After 5 years
                    count += 1
        equity_rounds_after_5_years.append(count)
    new_df['equity_rounds_raised_after_5_years'] = equity_rounds_after_5_years

    # Calculate Equity_raised_until_now
    total_equity_raised = []
    for index, row in df.iterrows():
        if pd.notna(row['acquisition_announced_on_date']):  # If acquisition date is not NaN
            total_equity_sum = sum([row[col] for col in equity_columns if pd.notna(row[col])])
        else:
            total_equity_sum = 0
        total_equity_raised.append(total_equity_sum)
    new_df['Equity_raised_until_now'] = total_equity_raised

    # Calculate Success_until_now
    new_df['Success_until_now'] = df['acquisition_announced_on_date'].notna().astype(int)

    return new_df

# Assuming your dataframe is Open_Startup_Total_df
new_dataframe = create_new_dataframe(Open_Startup_Total_df)
```

Filter if the founders actually for at this company: thanks to the usage of google lense we could extract founders, which LinkedIn otherwise would not give us the access to, because google lens would extract the founders, LinkedIn link according the Picture which LinkedIn would provide us. But there were also some missclassifications of google lens. These founder now have to be filtered out.

```python
import pandas as pd
from fuzzywuzzy import fuzz

# Function to check if the company name appears in the experience
def check_company_in_experience(row):
    try:
        experiences = eval(row['experience'])  # Parse the experience column
        company_name = row['Company Name'].lower()  # Normalize to lowercase
        for exp in experiences:
            exp_company = exp.get('company_name', '').lower()  # Extract and normalize company name
            if fuzz.partial_ratio(company_name, exp_company) >= 90:  # Check for 90% similarity
                return True
    except:
        pass
    return False

# Apply the function to filter rows
filtered_open_data_df = open_data_df[open_data_df.apply(check_company_in_experience, axis=1)]

# Display the filtered DataFrame
filtered_open_data_df
```
